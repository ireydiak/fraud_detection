## Débalancement de classe
Les données présentent un fort débalancement dans les classes. Certaines classes sont beaucoup plus nombreuses que d'autres.

```python
df["Class"].explode().value_counts()
```

En effet, la classe 2 possède seulement deux observations tandis que la classe -1 en possède plutôt 56. On utilise l'algorithme SMOTE pour ré-équilibrer le tout. On doit être prudent en utilisant cette technique car on peut créer des clusters artificiels qui ne reflètent pas nécessairement la réalité empirique. On ré-équilibre donc les classes -2 et 1 en fonction de la classe 0. Nous devons malheureusement ignorer la classe 2 car elle ne contient pas assez de données.

```python
sm = SMOTE(
    random_state=42,
    sampling_strategy={"Class 1": 48, "Class -2": 48}
)
X, y = df.loc[:, df.columns != "Class"].to_numpy(), df.loc[:, "Class"].to_numpy() 
X_ovsld, y_ovsld = sm.fit_resample(X, y)
X = np.concatenate((np.expand_dims(y_ovsld, 1), X_ovsld), axis=1)
df = pd.DataFrame(X, columns=df.columns)
df.head(5)
```

On confirme que les classes -2, 1 et 0 sont ré-équilibrées.

```python
df["Class"].explode().value_counts()
```

Comme mentionné précédemment, les variables respectent des échelles de grandeur très différentes.
Il faut donc les centrer et les réduire.

```python
from sklearn.preprocessing import StandardScaler

# On copie le dataframe (sans les étiquettes) dans une matrice numpy X
X = df.loc[:, df.columns != "Class"].to_numpy(dtype=np.float32)
y = df.loc[:, "Class"]
# On applique un centrage et une réduction par équart-type de la matrice X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])

# On valide la dernière étape en s'assurant que les moyennes et écart-types 
# soient égaux à 0 et 1 respectivement
assert np.allclose(0., X_scaled.mean(axis=0), atol=1e-6), "Moyennes non centrées"
assert np.allclose(1., X_scaled.std(axis=0), atol=1e-6), "Écart-types différents de 1"
```

### Sélection des 10 meilleurs attributs
Nous sélectionnons les 10 meilleurs attributs à l'aide d'une méthode ensembliste (ensemble methods). Cette technique n'a pas été étudiée en classe, mais elle est commune en machine learning. On combine la sortie de plusieurs algorithmes différents pour obtenir de meilleurs résultats. On considère quatre algorithmes différents: SelectKBest basé sur l'entropie différentielle, un arbre de décision et finalement les deux algorithmes de gradient boosting les plus populaires (XGBoost et Catboost). On calcule l'union des 10 attributs retenus par chaque méthode et on sélectionne les 10 attributs qui reviennent le plus fréquemment.

```python
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.tree import DecisionTreeClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from collections import Counter

# SelectKBest
clf = SelectKBest(mutual_info_classif, k=10)
clf.fit(df_scaled[num_cols], y)
features_a = list(clf.get_feature_names_out())

# DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion="entropy", random_state=42, max_features=10)
clf.fit(X_scaled, y)
ind = np.argpartition(clf.feature_importances_, -10)[-10:]
features_b = list(df.columns[ind])

# CatBoostClassifier
clf = CatBoostClassifier(verbose=False)
clf.fit(X_scaled, y)
ind = np.argpartition(clf.feature_importances_, -10)[-10:]
features_c = list(df.columns[ind])

# XGboost
clf = XGBClassifier()
clf.fit(X_scaled, y)
ind = np.argpartition(clf.feature_importances_, -10)[-10:]
features_d = list(df.columns[ind])

def default_value():
    return 0

all_features = features_a + features_b + features_c + features_d
features_dict = Counter(all_features)
features_df = pd.DataFrame(
    list(zip(features_dict.keys(), features_dict.values())),
    columns=["Attribut", "Décompte"]
).sort_values(ascending=False, by="Décompte")
features_df
```

Nous sommes chanceux et ne devons pas gérer le cas où le 11e attribut apparaît aussi souvent que le 10e. On peut finalement sauvegarder nos données.

```python
selected_features = features_df.loc[:, "Attribut"][0:10].tolist()
X = df_scaled.loc[:, selected_features].to_numpy(dtype=np.float32)
df_final = pd.DataFrame(
    df_scaled.loc[:, selected_features + ["Class"]],
    index=df_scaled.index,
    columns=selected_features + ["Class"]
)
df_final.to_csv("./QSAR_10best_dataset.csv", index=False)
df_final.loc[:, selected_features] = df_final.loc[:, selected_features].astype(float)
selected_features
```

## Analyse de la corrélation
Maintenant que les données sont prétraitées, on peut s'attarder à l'analyse des corrélations. Le premier graphique affiche les corrélations pour toutes les variables tandis que le second affiche les corrélations des attributs pour chacune des classe.

On constate que les données ne sont pas corrélées de la même manière en fonction des classes. Les variables sont fortement corrélées pour la classe -2. L'attribut vsurf_R demeure le plus consistant avec le moins de corrélation cumulative entre les classes. Seule la classe 2 présente une corrélation non négligeable pour cet attribut. La présence de corrélation rend l'analyse par composantes principales pertinente car il sera possible de trouver des axes préservant cette corrélation.

Nous procédons maintenant à une réduction de dimension par ACP. Nous n'utilisons pas la librairie SKLearn mais effectuons plutôt les calculs manuellement pour des fins éducatives.

```python
# Création de la matrice de corélation
Sigma = (1 / (len(df))) * (X.T @ X)
# Vérification que la diagonale de la matrice de corélation est contient des uns.
assert np.allclose(np.diag(Sigma), np.ones(Sigma.shape[0]), atol=1e-6)

# Décomposition valeurs, vecteurs propres
eig_vals, eig_vecs = np.linalg.eig(Sigma)
# Trier les vecteurs et valeurs propres de manière décroissante
# https://stackoverflow.com/questions/8092920/sort-eigenvalues-and-associated-eigenvectors-after-using-numpy-linalg-eig-in-pyt
idx = eig_vals.argsort()[::-1]   
eig_vals = eig_vals[idx]
eig_vecs = eig_vecs[:, idx]

# D'abord on calcule la somme de chaque valeur propre
sum_eigval = np.sum(eig_vals)
# Ensuite, on divise chacune des valeur propre par cette somme
var_prop = eig_vals / sum_eigval
cum_prop = np.cumsum(var_prop)
# Validation que la somme des proportions donne bien 1
assert(np.allclose(var_prop.sum(), 1))

# On crée une table pour présenter les résultats
d = {"Valeur propre": eig_vals, "Pourcentages": np.around(var_prop * 100, 4), "Pourcentages cumulés": np.around(cum_prop * 100, 4)}
eig_df = pd.DataFrame(data=d)
eig_df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

x_labels = [i + 1 for i in range(len(var_prop))]

ax.plot(x_labels[0:15], var_prop[0:15], marker='o', markersize=6, color='blue', linewidth=2, label="Proportion de la variance")
ax.plot(x_labels[0:15], cum_prop[0:15], marker='o', color='orange', linewidth=2, label="Variance cumulée")
ax.legend()
ax.set_xticks(x_labels[0:15])
ax.set_title("Diagramme d'éboulis")
ax.set_xlabel("Composante principale")
ax.set_ylabel("Proportion de variance")
plt.show()
```

Le diagramme d'éboulis ainsi que le tableau des composantes principales nous informent que 86% de la corrélation est expliquée par seulement
les trois premières composantes principales. On projette donc nos données sur ces trois composantes. Il y a une différence de 10% entre la deuxième et la troisième composante. Par conséquent, la projection 3CP est favorisée quoique 2CP suffirait probablement avec 76% de corrélation variance cumulée. On évalue du même coup la qualité de la représentation à l'aide de deux métriques.
- $cos(\hat{X}, X) = \frac{||\hat{X}||^2}{||X||^2}$
- $d^2(\hat{X}, X) = ||\hat{X} - X||^2$

```python
M = 3
Pi = eig_vecs[:, 0:M]
Z = X @ Pi
X_hat = Z @ Pi.T
assert X_hat.shape == X.shape, "mauvais format de matrice"

d_cos = np.linalg.norm(X_hat)**2 / np.linalg.norm(X)**2
d_euc = np.linalg.norm(X_hat - X)
print("d_cos = {:.4f}\nd_euc = {:.4f}".format(d_cos, d_euc))
```

L'angle entre l'espace latent et les variable originales est de l'ordre de 0.9063, ce qui nous indique que la projection est très fidèle et ne perd pas beaucoup d'information.

Une partie intéressante de l'analyse par composantes principales est de calculer la corrélation entre les variables initiales et les composantes principales. En affichant ces données dans le cercle des corrélations, une personne experte dans le domaine donner une signification aux axes.

```python
eig_vals[eig_vals <= 0] = 0
L = eig_vecs * np.sqrt(eig_vals)
loading_df = pd.DataFrame(
    L[:, 0:M],
    index=selected_features,
    columns=["CP-{}".format(i + 1) for i in range(len(eig_vecs[0:M]))]
)
loading_df.head(5)
```

## Calcul des centroids
Pour confirmer le choix entre 3CP et 2CP, on vérifie les performances d'un classifieur. Ce classifieur assigne à une observation la classe du centroid dont il est le plus proche. Afficher centroids (représentés par moyenne arithémtique) de chaque composante principale pour chaque classe. On commence par afficher les centroids - représentés par la moyenne arithmétique - de chaque composante principale pour chacune des classes.

```python
from sklearn.preprocessing import LabelEncoder
from models import NearestCentroid

y_true = df_acp.loc[:, "Class"].to_numpy()
le = LabelEncoder()
y_true_enc = le.fit_transform(y_true)
```

```python
df_centroids = df_acp.groupby("Class").mean()
acp_num_cols = df_acp.select_dtypes(exclude=object).columns.tolist()
df_centroids
```

On peut maintenant entraîner le classifieur sur les données.

```python
def mahalanobis(X: np.array, mu: np.array, S: np.array):
    S_inv = np.linalg.inv(S)
    return (X - mu).T @ S_inv @ (X - mu)

def nearest_centroid_classifier(data: np.array, targets, label_enc, dist_measure="euclidean"):
    clf = NearestCentroid(label_encoder=label_enc, metric=dist_measure)
    clf.fit(data, targets)
    y_pred = clf.predict(data)
    return clf.evaluate(targets, y_pred)
```

```python
# Transfert des données réduites dans une liste numpy
data = df_acp.loc[:, acp_num_cols].to_numpy()

# Entraînement du classifieur sur les données représentées respectivement par 
# deux et trois composantes principales
df_metrics_3cp = nearest_centroid_classifier(data, y_true, le)
df_metrics_2cp = nearest_centroid_classifier(data[:, 0:2], y_true, le)
f, axs = plt.subplots(1, 3, figsize=(24, 12))

# On sauvegarde les métriques de performance dans des DataFrame
df_precision = pd.concat([
    df_metrics_2cp.loc["precision"][0:5].rename("2-CP"), 
    df_metrics_3cp.loc["precision"][0:5].rename("3-CP")
], axis=1)
df_recall = pd.concat([
    df_metrics_2cp.loc["recall"][0:5].rename("2-CP"), 
    df_metrics_3cp.loc["recall"][0:5].rename("3-CP")
], axis=1)
df_f1 = pd.concat([
    df_metrics_2cp.loc["f1-score"][0:5].rename("2-CP"), 
    df_metrics_3cp.loc["f1-score"][0:5].rename("3-CP")
], axis=1)

# Affichage
df_precision.plot(kind="barh", ax=axs[0])
df_recall.plot(kind="barh", ax=axs[1])
df_f1.plot(kind="barh", ax=axs[2])

axs[0].set_title("Precision")
axs[0].set_xticks(np.arange(0, 1, step=0.1))
axs[1].set_title("Recall")
axs[2].set_title("F1-score")

plt.show()
```

On constate que la precision, recall (et par extension le f1-score) sont systématiquement supérieurs pour le cas 3CP.
