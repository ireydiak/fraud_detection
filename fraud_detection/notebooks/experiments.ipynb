```python
import os.path
from anomaly_detection import bootstrap

dataset_files = [os.path.abspath("../data/fraud-detection-v%d.npy") % i for i in range(1, 5)]
dataset_files
```

```python
n_epochs = 200
batch_size = 128
n_runs = 10
lr = 1e-03

for dataset_path in dataset_files:
    for model_name in ["DeepSVDD", "DAGMM"]:
        bootstrap.train(
            model_name=model_name,
            dataset_path=dataset_path,
            batch_size=batch_size,
            n_runs=n_runs,
            n_epochs=n_epochs,
            learning_rate=lr,
            results_path=os.path.abspath("../results"),
            models_path=os.path.abspath("../models"),
            test_mode=False
        )
```
